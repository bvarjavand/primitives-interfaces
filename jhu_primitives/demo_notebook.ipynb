```python
import numpy as np

from jhu_primitives.utils.util import gen_graph_r
from jhu_primitives.core.JHUGraph import JHUGraph
from jhu_primitives import *
import matplotlib
import matplotlib.pyplot as plt
```

```python
gpath, rig = gen_graph_r(n=50, p=.1)

g = JHUGraph()
g.read_graph(fname=gpath)

gmtx = np.array(g.get_dense_matrix())
fig = plt.figure()
ax = fig.add_subplot(111)
ax.matshow(gmtx)
ax.set_ylabel('vertex id')
ax.set_xlabel('vertex id')
ax.set_title('ER Random graph with n=50 and p=.1')
fig.set_size_inches(8, 6)
plt.colorbar
plt.show()

lmtx = gmtx - np.diag(gmtx.sum(axis=1))
fig = plt.figure()
ax = fig.add_subplot(111)
ax.matshow(lmtx)
ax.set_ylabel('vertex id')
ax.set_xlabel('vertex id')
ax.set_title('Laplacian Matrix of ER Random Graph')
fig.set_size_inches(8, 6)
plt.colorbar
plt.show()
```

```python
print("Summary: ")
g.summary()

ASE = AdjacencySpectralEmbedding()
ase_embed = ASE.embed(g=g, dim=4)

fig = plt.figure()
ax = fig.add_subplot(111)
ax.matshow(ase_embed)
ax.set_xlabel('spectral dimension')
ax.set_ylabel('vertex id')
ax.set_title('First 4 Adjacency Spectral Dimensions')
fig.set_size_inches(8, 6)
plt.colorbar
plt.show()
```

```python
LSE = LaplacianSpectralEmbedding()
lse_embed = LSE.embed(g=g, dim=4)

fig = plt.figure()
ax = fig.add_subplot(111)
ax.matshow(lse_embed)
ax.set_xlabel('spectral dimension')
ax.set_ylabel('vertex id')
ax.set_title('First 4 Laplacian Spectral Dimensions')
fig.set_size_inches(8, 6)
plt.colorbar
plt.show()
```

```python
GCLUST = GaussianClustering()

random_ar = np.random.random((64, 8))
fig = plt.figure()
ax = fig.add_subplot(111)
ax.matshow(np.array(random_ar))
ax.set_xlabel('Features')
ax.set_ylabel('Samples')
ax.set_title('Random Entry Data')
fig.set_size_inches(12, 8)
plt.colorbar
plt.show()

print("GCLUST: ",
        GCLUST.cluster(inputs=random_ar, dim=4), "\n\n")
print("GCLUST: ",
        GCLUST.produce(inputs=random_ar), "\n\n")
```

```python
NONPAR = NonParametricClustering()
xhat1 = np.random.random((16, 2))
xhat2 = np.random.random((16, 2))

fig = plt.figure()
ax = fig.add_subplot(111)
ax.matshow(xhat1)
ax.set_ylabel('Sample')
ax.set_xlabel('Feature')
ax.set_title('Xhat1')
plt.show()

fig = plt.figure()
ax = fig.add_subplot(111)
ax.matshow(xhat2)
ax.set_ylabel('Sample')
ax.set_xlabel('Feature')
ax.set_title('Xhat2')
plt.show()

npclust = NONPAR.cluster(xhat1=xhat1, xhat2=xhat2)
print("NONPAR: ", npclust, "\n\n")

# OOCASE = OutOfCoreAdjacencySpectralEmbedding()
# print("OOCASE: ", OOCASE.embed(g=g), "\n\n")
```

```python
gpath, rig = gen_graph_r(n=50, p=.1)
g2 = JHUGraph()
g2.read_graph(fname=gpath)

fig = plt.figure()
ax = fig.add_subplot(111)
ax.matshow(g.get_dense_matrix())
ax.set_ylabel('vertex id')
ax.set_xlabel('vertex  id')
ax.set_title('First Random Graph')
plt.show()


fig = plt.figure()
ax = fig.add_subplot(111)
ax.matshow(g2.get_dense_matrix())
ax.set_ylabel('vertex id')
ax.set_xlabel('vertex  id')
ax.set_title('Second Random Graph')
plt.show()

SGM = SeededGraphMatching()
sgm_out = SGM.match(g1=g, g2=g2)
print(sgm_out)
print("SGM: ", sgm_out, "\n\n")
```
